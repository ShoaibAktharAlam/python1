## Accept a number from the user and check whether it's a positive number or not.

```python
n=int(input("Enter a number:"))
if(n>0):
    print("The given number is positive")
elif(n<0):
    print("The given number is negative")
else:
    print("The number is zero")


```

## Accept three sides of the triangle and check whether the triangle is possible or not.

```python
n1=int(input("Enter side.1:"))
n2=int(input("Enter side.2:"))
n3=int(input("Enter side.3:"))
if(n1+n2>n3 and n2+n3>n1 and n1+n3>n2):
    print("The tringle is possible")
else:
    print("The tringle is not possible")
```
